```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import SGDRegressor
```

### Reponse 1: generate  2 vectorrs x and y :
generate two random vectors x and y wuth shape of (100,1)(100 rows and one columns)

```python
x = 2*np.random.rand(100,1)
y = 4+3*x+np.random.randn(100,1)*0.5
```

## reponse 2: Visualize the data with plt.scatter
plt.scatter(X,y):  creating scatter plots eatch data point represented with a dot

plt.show():display the scatter

```python
## Reponse 2
```

```python
plt.scatter(x,y)
plt.xlabel('X_values')
plt.ylabel('Y_values')
plt.show()
```

## Reponse 3:
SGDRegressor create a  Stochastic Gradient Descent (SGD) with 100 iterations and small learning rate 0.0001 to  avoid overfitting :
model.fit to training the model

```python
model = SGDRegressor(max_iter=100,eta0=0.0001)
model.fit(x,y)
```

## reponse 4: Calculate Accuracy :
calculate the cofficient of determination usimg model.score or r2_score

```python
from sklearn.metrics import r2_score
r2 = model.score(x,y)
print(f"the accuracy is :{r2}")
```

## reponse 5 : interpret the resault
The negative R-squared value of -9.314908207188688 indicates a very poor model fit.a low learning rate causes the model to converge slowly because it takes smaller steps in adjusting the weights during training. This gradual approach can prevent the model from overshooting the optimal solution, but it also means it requires more iterations to reach a good minimum. When combined with a small number of epochs, this effect is amplified—the model doesn’t have enough training iterations to make substantial progress toward the optimal solution. As a result, the model may get "stuck" far from the minimum, yielding higher loss and potentially underperforming

## the learning curve:


```python
from io import StringIO
import sys
old_stdout = sys.stdout
sys.stdout = mystdout = StringIO()
clf = SGDRegressor(max_iter= 1000, alpha= 0.0001, verbose=1)
clf.fit(x,y.ravel())
sys.stdout = old_stdout
loss_history = mystdout.getvalue()
loss_list = []
for line in loss_history.split('\n'):
    if(len(line.split("loss: ")) == 1):
        continue
    loss_list.append(float(line.split("loss: ")[-1]))
plt.figure()
plt.plot(np.arange(len(loss_list)), loss_list)
plt.xlabel("Time in epochs")
plt.ylabel("Loss")
```

In the first case, the model has a small number of epochs and a low learning rate, causing it to converge slowly toward the minimum. However, in the second case, with a larger number of epochs and a higher learning rate, the model converges more quickly. As we observe, the higher learning rate leads to a lower loss after only 5 epochs.
